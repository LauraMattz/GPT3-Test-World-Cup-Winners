##  GPT3 Test World Cup Winners

  Prompt: Create a graphic using python with the sum of the winners of world cups and put in order of wins and a different color for each winner 

```python
import matplotlib.pyplot as plt
import pandas as pd

# define the data as a list of tuples, where the first element is the year and the second element is the winner
data = [
    (1930, "Uruguay"),
    (1934, "Italy"),
    (1938, "Italy"),
    (1950, "Uruguay"),
    (1954, "West Germany"),
    (1958, "Brazil"),
    (1962, "Brazil"),
    (1966, "England"),
    (1970, "Brazil"),
    (1974, "West Germany"),
    (1978, "Argentina"),
    (1982, "Italy"),
    (1986, "Argentina"),
    (1990, "West Germany"),
    (1994, "Brazil"),
    (1998, "France"),
    (2002, "Brazil"),
    (2006, "Italy"),
    (2010, "Spain"),
    (2014, "Germany"),
    (2018, "France")
]

# create a DataFrame from the data
df = pd.DataFrame(data, columns=["year", "winner"])

# group the data by winner and count the number of wins for each team
win_count = df.groupby("winner").size().reset_index(name="wins")

# sort the data by the number of wins in descending order
win_count = win_count.sort_values(by="wins", ascending=False)

# define a list of colors for the bars
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b", "#e377c2", "#7f7f7f", "#bcbd22", "#17becf"]

# create the bar chart
plt.figure(figsize=(12, 8))  # specify the size of the plot
plt.bar(win_count["winner"], win_count["wins"], color=colors)

# customize the plot
plt.title("Number of World Cup Wins", fontsize=20)
plt.xlabel("Winner", fontsize=16)
plt.ylabel("Number of Wins", fontsize=16)
plt.xticks(rotation=45, fontsize=14)
plt.yticks(fontsize=14)

# show the plot
plt.show()
```
